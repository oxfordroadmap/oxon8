### 主要指标如 ``二氧化碳排放`` 及 ``碳排強度`` 要降多少？

主要指标如 ``二氧化碳排放`` 及 ``碳排強度`` 要降多少？

PwC 及 澳恪森研究均指出，世界碳排强度需``年均减少17-19%``，才能达《巴黎协定》1.5℃目标。

中国2024年[《2024—2025年节能降碳行动方案》](https://www.gov.cn/zhengce/202405/content_6954583.htm)要求碳排强度（经济） ``年减3.9%``。

此表交互可以选定指标间的比较。

:::{admonition} ``二氧化碳排放`` 指标要降多少？
:class: important

* 中国2024年已下指标要``年减3.9%``左右（``单位国内生产总值二氧化碳排放``，见[《2024—2025年节能降碳行动方案》](https://www.gov.cn/zhengce/202405/content_6954583.htm)，可视为``经济年脱碳率``或``经济碳排强度年減率``）
* PwC指出，世界碳排强度需``年均减少17.2%``，才能达《巴黎协定》1.5℃目标。见[《2023淨零排放經濟指數報告》（Net Zero Economy Index 2023）](https://www.pwc.tw/zh/news/press-release/press-20231011.html)。
* 澳恪森研究发现，世界碳排强度需``年均减少19%``，才能达1.5℃目标（见[VisNetZero成果](https://oxon8.netlify.app/visualization/prj-visnetzero/index.zh)）。[包括中国的新兴七国，2030年前需加4倍努力，达到年减9.8-10.3% 的脱碳率](https://oxon8.netlify.app/visualization/prj-visnetzero/netzero.co2pg.e7.zh)，才能达1.5℃目标。近年中国脱碳率表现最好的年份及数据分别为 -7.773%（2015年）、 -7.334%（2016年）、 -6.822%（1980年）、 -6.495%（2014年）、等。最近2021年亦达-4.180%。中国2024年方案指标相比之下并不严苛。

:::
