### 桑基图的重要性

桑基图对探究一经济体的 ``产业结构`` 和 ``能源结构``（Energy Mix）的交互关系至关重要。

#### 投入产出表的桑基图可提供关系洞察

投入产出表的桑基图可提供以下洞察：

* 该经济体的 ``主要行业部门`` 的投入产出关系
* 该经济体的 ``能源`` 及 ``非能源`` 行业部门的投入产出关系


`````{admonition} 研究笔记：投入产出表的桑基图可回流
:class: note
不同于能源的桑基图（单向，不能回流），投入产出表的桑基图有回流的可能。本桑基图按行业从自然资源到社会消费的流动顺序假设出发，对行业部门先进行排序后，并再将回流的关系标出，以配合桑基图不可回流的特性。
`````
#### 传统桑基图用于能源供应与消费

桑基图（Sankey）常用于 ``能源供应与消费`` 是否平衡、是否有效率等等的管理与决策。

![../figures/ENERGY_2017_CHINA.png](../figures/ENERGY_2017_CHINA.png)
2017年中国能源供应与消费图

国际能源署（IEA）为世界各经济体制作桑基图，展示能源从源到端的生产、进口、加工、分布、及消耗情况。桑基图也能展示能源 ``能源结构``（Energy Mix）。

#### 小结

投入产出表的桑基图，可搭配能源的情报来探究一经济体的 ``产业结构`` 和 ``能源结构``（Energy Mix）的交互关系，透过理解其关系的变化及消长，来探究 ``净零经济`` 的 ``产业转型`` 及 ``能源转型`` 方案。
