### 哪些行业部门需要重点关注？

一般来说，重点关注高耗能及高排放的行业部门，以及其上下游的 ``产业链`` ，而 ``投入产出`` 表则提供需关注的 ``产业关系``。

#### 2024年中国《行动方案》

《行动方案》围绕能源、工业、建筑、交通、公共机构、用能设备等重点领域和重点行业，部署了节能降碳十大行动。

:::{admonition} VisCEADs 是聚焦重点领域和重点行业好帮手
:class: important

1. 投入产出值构成的 ``产业关系`` 可视化，能将重点领域和重点行业一目瞭然，并追踪其上下游及历史走向。
2. CEADs 数据的 ``分省分行业数据粒度``，可为省级及行业领导或决策者提供数据驱动基础， ``精准脱碳``。
3. 此表提供的相关进出口、就业、等信息，可用来丰富精准脱碳的政策品质，精准脱碳 ``同时可达其它政策需求``如 ``稳民生或促出口``。
:::
